### Load 'Depression by education level & employment status' data. 
This data is based on adults aged 25-64 years old, and focus on self-reported depression only. The lack of differentiation in these surveys between mental health disorders, such as depression, anxiety disorders, and bipolar disorder mean that self-reported depression data may include individuals with these other disorders.

These are additional explanations for variables.

* employed: whether groups are employed
* active: actively seeking employment
* total: the total of employed, active and unemployed. 
* tertiary (postsecondary) education: refers to any type of education pursued beyond the high school level. 
* upper secondary education: comprises three years of schooling (grades 10, 11 and 12) and is the stage before tertiary education.
* Post-secondary non-tertiary education: provides learning and educational activities building on secondary education preparing for both labour market entry as well as tertiary education. Programmes are often not significantly more advanced than those at upper secondary as they typically serve to broaden rather than deepen knowledge, skills and competencies.

Information above is from https://ourworldindata.org/mental-health#all-charts-preview.

# Combine interested data

Interested data list:
* df1_age
* df1_gender
* df1_suicide
* df1_number
* df2_unemployment
* df2_gdp

```python
# convert columns to rows
df2_unemployment_2 = pd.melt(df2_unemployment, id_vars=["Country Name", 'Country Code'], var_name="Year", value_name="UnemploymentRate(%)")
```

```python
df2_unemployment_2 = df2_unemployment_2.sort_values(by=["Country Name", 'Year']).reset_index(drop=True)
df2_unemployment_2.head()
```

```python
df2_gdp_2 = pd.melt(df2_gdp, id_vars=["Country Name", 'Country Code'], var_name="Year", value_name="GDP_per_capita($)")
df2_gdp_2 = df2_gdp_2.sort_values(by=["Country Name", 'Year']).reset_index(drop=True)
df2_gdp_2.head()
```

```python
df2_unemployment_2.set_index(['Country Name', 'Country Code', 'Year'], inplace=True)
df2_gdp_2.set_index(['Country Name','Country Code', 'Year'], inplace=True)
```

```python
df2_unemployment_2.head()
```

```python
df2_gdp_2.head()
```

```python
# Change the name of column to be the same
df1_age.rename(columns={'Entity':'Country Name', 'Code': 'Country Code'}, inplace=True)
df1_gender.rename(columns={'Entity':'Country Name', 'Code': 'Country Code'}, inplace=True)
df1_suicide.rename(columns={'Entity':'Country Name', 'Code': 'Country Code'}, inplace=True)
df1_number.rename(columns={'Entity':'Country Name', 'Code': 'Country Code'}, inplace=True)
```

```python
df1_age.set_index(['Country Name', 'Country Code', 'Year'], inplace=True)
df1_gender.set_index(['Country Name', 'Country Code','Year'], inplace=True)
df1_suicide.set_index(['Country Name', 'Country Code','Year'], inplace=True)
df1_number.set_index(['Country Name', 'Country Code','Year'], inplace=True)
```

```python
# Merge those dataframes
data_frames = [df1_age, df1_gender, df1_suicide, df1_number, df2_unemployment_2, df2_gdp_2]
df_merged = pd.concat(data_frames, join='outer', axis=1).fillna(np.nan)
```

```python
df_merged.head()
```

```python
df_merged.reset_index(inplace=True)
df_merged.head(100)
```
