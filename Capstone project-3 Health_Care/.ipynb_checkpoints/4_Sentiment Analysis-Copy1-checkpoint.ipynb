# Sentiment Analysis of Health Tweets -- Sentiment Anlysis

## Introduction

In this notebook, I will apply the TextBlob function from textblob library to do the sentiment analysis for each tweets. This method will give a polarity score for each tweet. Then we can label tweets based on polarity score to **positive**, **negative** or **neutral**. 

**Polarity**: How positive or negative a word is. -1 is very negative. +1 is very positive. 

Then I checked the tweets precentage of each sentiment category and the most common words and health words in each sentiment category.

## Loading the data

```python
#load python packages
import os
import pandas as pd
import datetime
import time
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
import string
from datetime import datetime
from nltk.corpus import stopwords
import nltk
from collections import Counter
from wordcloud import WordCloud

from textblob import TextBlob
%matplotlib inline
```

```python
df = pd.read_csv('data/clean_data_2.csv',index_col=0)
df.head()
```

## Sentiment of tweets

### Getting the polarity scores for each tweet

```python
# Create quick lambda functions to find the polarity of each tweet

pol = lambda x: TextBlob(x).sentiment.polarity

df['polarity'] = df['text'].apply(pol)
df.head()
```

### Labeling tweets based on the polarity value

```python
df['label'] = ['positive' if score > 0 else 'negative' if score < 0 else 'neutral' for score in df.polarity]
df.head()
```

```python
data = df.copy()
```

### Sentiment counts

```python
counts_df = data.label.value_counts().reset_index()
counts_df.columns = ['sentiment', 'count']
counts_df
```

```python
plt.title("Sentiment count", fontsize = 18)
sns.barplot(x='sentiment', y='count', data=counts_df)
plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\bar_sentiment_count.png')
```

We observe that most health tweets are positive.

### WordCloud for each sentiment

```python
# filter negative data
df_neg = data[data.label == 'negative']
# Get all words in negative tweets
words_list_neg = [word for line in df_neg.text for word in line.split()]
# plot wordcloud for negative tweets
wc = WordCloud(width=800, height=500, random_state=42)
plt.figure(figsize=(10, 7))
wc.generate(' '.join(words_list_neg))
plt.imshow(wc, interpolation="bilinear")
plt.title("Vocabulary from Negative Tweets", fontsize = 22)
plt.axis("off")
plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\wordcloud_negative.png')
```

Most common words in negative tweets are 'sick', 'cold', 'sore', 'anxious', 'ill' and some bad words, etc...

```python
df_pos = data[data.label == 'positive']
words_list_pos = [word for line in df_pos.text for word in line.split()]

wc = WordCloud(width=800, height=500, random_state=42)
plt.figure(figsize=(10, 7))
wc.generate(' '.join(words_list_pos))
plt.imshow(wc, interpolation="bilinear")
plt.title("Vocabulary from Positive Tweets", fontsize = 22)
plt.axis("off")
plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\wordcloud_positive.png')
```

Most common words in positive tweets are 'good', 'well', 'think', 'lol', and 'doctor', etc...

```python
df_neu = data[data.label == 'neutral']
words_list_neu = [word for line in df_neu.text for word in line.split()]

wc = WordCloud(width=800, height=500, random_state=42)
plt.figure(figsize=(10, 7))
wc.generate(' '.join(words_list_neu))
plt.imshow(wc, interpolation="bilinear")
plt.title("Vocabulary from Neutral Tweets", fontsize = 22)
plt.axis("off")
plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\wordcloud_neutral.png')
```

Most common words in neutral tweets are 'pain', 'sore', 'body', 'anxiety', 'killing', and 'loser', etc... But they are all in the same frequency level.

### Tweets's length and word counts in each sentiment

```python
#Calculating tweetâ€™s length and word count
data['text_len'] = data['text'].astype(str).apply(len)
data['text_word_count'] = data['text'].apply(lambda x: len(str(x).split()))
df_length = round(pd.DataFrame(data.groupby('label').text_len.mean()),2)
```

```python
df_word_count = round(pd.DataFrame(data.groupby('label').text_word_count.mean()),2)
```

```python
plt.title("Tweet's length for each sentiment", fontsize = 18)
sns.barplot(x=df_length.index, y='text_len', data=df_length)
plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\bar_sentiment_length.png')
```

```python
plt.title("Tweet's words count for each sentiment", fontsize = 18)
sns.barplot(x=df_word_count.index, y='text_word_count', data=df_word_count)
plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\bar_sentiment_word_counts.png')
```

In positive tweets, people tend to write more. People tend to write less in negative tweets and least in neutral tweets. 

### Frequency of health words in each sentiment category

```python
# read the keywords from txt to a list
txt_list = []

with open('keywords.txt', "r") as f:
    txt_list = f.read().split()

keywords = [word.strip(',') for word in txt_list]

# build health words list for each sentiment categoty
words_list_keywords_neg = [word for word in words_list_neg if word in keywords]
words_list_keywords_pos = [word for word in words_list_pos if word in keywords]
words_list_keywords_neu = [word for word in words_list_neu if word in keywords]

# build health words count for each sentiment categoty
word_counts_keywords_neg = Counter(words_list_keywords_neg).most_common()
neg_keywords_df = pd.DataFrame(word_counts_keywords_neg)
neg_keywords_df.columns = ['word', 'frequency']
neg_keywords_df = neg_keywords_df.sort_values('frequency', ascending=False)

word_counts_keywords_pos = Counter(words_list_keywords_pos).most_common()
pos_keywords_df = pd.DataFrame(word_counts_keywords_pos)
pos_keywords_df.columns = ['word', 'frequency']
pos_keywords_df = pos_keywords_df.sort_values('frequency', ascending=False)

word_counts_keywords_neu = Counter(words_list_keywords_neu).most_common()
neu_keywords_df = pd.DataFrame(word_counts_keywords_neu)
neu_keywords_df.columns = ['word', 'frequency']
neu_keywords_df = neu_keywords_df.sort_values('frequency', ascending=False)
```

#### Plot the most frequency words in each sentiment category

```python
plt.figure(figsize=(16, 6))
plt.subplot(1,3,1)
sns.barplot(x='frequency', y='word', data=neg_keywords_df[:20], palette="Blues_d")
plt.xlim([0, 8000])
plt.title("Health words frequency in negative tweets", fontsize = 10)

plt.subplot(1,3,2)
sns.barplot(x='frequency', y='word', data=pos_keywords_df[:20], palette="Blues_d")
plt.xlim([0, 8000])
plt.title("Health words frequency in positive tweets", fontsize = 10)

plt.subplot(1,3,3)
sns.barplot(x='frequency', y='word', data=neu_keywords_df[:20], palette="Blues_d")
plt.xlim([0, 8000])
plt.title("Health words frequency in neutral tweets", fontsize = 10)

plt.tight_layout()

plt.savefig(r'C:\Users\yuhan\Desktop\Springboard\Capstone project-3 Health_Care\figures\bar_keywords_frequency_sentiment.png')
```

Here, we can observe that the frequency of health words in negative tweets much higher than in positive and neutral tweets.
And the most frequency health words in the negetive tweets and not in positive and neutral tweets are 'sick', 'ill', 'cold', 'anxious', 'exhausted', 'miserable', 'uncomfortable'.
